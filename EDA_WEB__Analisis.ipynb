# EDA: PAGINA WEB

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/drive/MyDrive/1 Bootcamp_Python_DSRP/8_Octava_Clase/web_1.xlsx")

df.head(5)
```

```python
#Cambiar de nombre las columnas:
df = df.rename(columns={'dia':'Fecha'})
df = df.rename(columns={'Visitas de la P치gina':'visita_pagina'})
df = df.rename(columns={'Tiempo promedio en la p치gina (segundos)':'tiempo_seg'})

```

```python
#df.head(5)
df.info()
```

```python
df['a침o'] = df['Fecha'].dt.year
df['mes'] = df['Fecha'].dt.month
df['mes_nombre'] = df['Fecha'].dt.month_name()
df['dia'] = df['Fecha'].dt.day
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df_filtered = df[df['a침o'].isin([2021,2022])]

df_filtered.tail(5)
```

```python
sns.histplot(
    data=df_filtered['visita_pagina'],
    bins=10,
    kde=False
)
plt.show()
```

```python
sns.histplot(
    data=df_filtered['tiempo_seg'],
    bins=10,
    kde=False
)
plt.show()
```

```python
sns.histplot(
    data=df_filtered['Clicks'],
    bins=10,
    kde=False
)
plt.show()
```

```python
sns.scatterplot(data=df_filtered, x='visita_pagina', y='tiempo_seg')
plt.title('visita_pagina vs tiempo_seg')
plt.xlabel('Visitas a la P치gina')
plt.ylabel('Tiempo Promedio en Segundos')
plt.show()
```

```python
sns.scatterplot(data=df_filtered, x='tiempo_seg', y='Clicks')
plt.title('tiempo_seg vs Clicks')
plt.xlabel('Tiempo Promedio en Segundos')
plt.ylabel('Clicks')
plt.show()
```

```python
df2 = pd.read_excel("/content/drive/MyDrive/1 Bootcamp_Python_DSRP/8_Octava_Clase/web_2.xlsx")

df2.head(5)
```

```python
#Sacando el margen y la rentabilidad:
df2['Margen'] =df2['Ingresos']- df2['Gasto']
df2['Rentabilidad'] =(df2['Margen']/ df2['Gasto'])*100

df2.head(5)
```

```python
df2 = df2.rename(columns={'dia':'Fecha'})

df2['a침o'] = df2['Fecha'].dt.year
df2['mes'] = df2['Fecha'].dt.month
df2['mes_nombre'] = df2['Fecha'].dt.month_name()
df2['dia'] = df2['Fecha'].dt.day
```

```python
df2.head(5)
```

```python
#Margen por a침o:

df2_filtered = df2[df2['a침o'].isin([2021,2022])]

df2_filtered.tail(4)


```

```python
#Margen
margen = df2_filtered.groupby('a침o')['Margen'].sum()
print(margen)
```

```python
#Rentabilidad
rent = df2_filtered.groupby('a침o')['Rentabilidad'].mean()

rent
```
